##  L'apprentissage automatique (machine learning) ,apprentissage artificiel ou apprentissage statistique est un champ d'étude de l'intelligence artificielle qui se fonde sur des approches mathématiques et statistiques pour donner aux ordinateurs la capacité d'« apprendre » à partir de données, c'est-à-dire d'améliorer leurs performances à résoudre des tâches sans être explicitement programmés pour chacune. ### Tic-Tac-Toe est un jeu simple pour deux joueurs auquel nous aimions jouer quand nous étions enfants (surtout dans les salles de classe ennuyeuses). Le jeu implique que 2 joueurs placent leurs symboles respectifs dans une grille 3x3. Le joueur qui parvient à placer trois de ses symboles en rangée horizontale/verticale/diagonale remporte la partie. Si l'un des joueurs ne parvient pas à le faire, la partie se termine par un match nul. Si les deux joueurs jouent toujours leurs stratégies optimales, le jeu se termine toujours par un match nul.

# Description des données
## Sources de données: https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/

## Dans cet dataset, nous avons 10 colonnes (9 pour les attributs et 1 pour les classes)
##  b,o,b,x,o,x,x,o,b => negative (partie nulle)
##  x,b,x,o,x,o,x,b,o => positive (P1 ou P2 gagnant)

```python
df = pd.read_csv('./tic-tac-toe.data',',')
```

```python
df.head()
```

## la nouvelle représentation du dataset est la suivante : 
## x => 2
## o => 1
## b => 0
## positive => 1
## negative => 0

## Cette méthode est utile pour obtenir une représentation numérique d'un tableau lorsque tout ce qui compte est d'identifier des valeurs distinctes.

```python
df['x'],v1 = pd.factorize(df['x'], sort=True)
df['x.1'],v2 = pd.factorize(df['x.1'], sort=True)
df['x.2'],v3 = pd.factorize(df['x.2'], sort=True)
df['x.3'],v4 = pd.factorize(df['x.3'], sort=True)
df['o'],v5 = pd.factorize(df['o'], sort=True)
df['o.1'],v6 = pd.factorize(df['o.1'], sort=True)
df['x.4'],v7 = pd.factorize(df['x.4'], sort=True)
df['o.2'],v8 = pd.factorize(df['o.2'], sort=True)
df['o.3'],v9 = pd.factorize(df['o.3'], sort=True)
df['positive'],v10 = pd.factorize(df['positive'], sort=True)
```

```python
df.info()
df.describe()
```

## Les attributs sont les 9 premiers colonnes

```python
feature_names = ['x','x.1','x.2','x.3', 'o', 'o.1', 'x.4', 'o.2', 'o.3']
```

```python
x_train = df[feature_names] # Features
x_train
```

## Les classes sont : La dérniere colonne 

```python
y_train = df['positive'] # Target
y_train
```

## Le gain d'informations est basé sur la diminution de l'entropie après la division d'un ensemble de données sur un attribut. Construire un arbre de décision consiste à trouver l'attribut qui renvoie le gain d'informations le plus élevé (c'est-à-dire les branches les plus homogènes).
## fit() est implémenté par chaque estimateur et il accepte une entrée pour les données d'échantillon ( X ) et pour les modèles supervisés, il accepte également un argument pour les étiquettes (c'est-à-dire les données cibles y ).
## plot_tree() pour Tracez un arbre de décision.
## Le graphe résultat est l'arbre compléte aprés l'entrainement du Modéle.

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(x_train, y_train)
clf
tree.plot_tree(clf)
plt.show()
```

## Les tests ont été effectués en recréant l'arbre de décision formé uniquement sur 70 % de l'ensemble de données et en le testant par rapport aux 30 % restants. Cela a été fait en utilisant la fonction train_test_split importée au début de ce cahier, et la méthode .predict() du DecisionTreeClassifier.

```python
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.3, random_state=1)
[x_train, x_test, y_train, y_test]
```

```python
clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=80) # change this classifier and check the impact
clf = clf.fit(x_train,y_train)
tree.plot_tree(clf)
plt.show()
```

```python
# use the model to make predictions with the test data
y_pred = clf.predict(x_test)
# how did our model perform?
count_misclassified = (y_test != y_pred).sum()
print('Les lignes non classés: {}'.format(count_misclassified))
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f} %'.format(accuracy * 100))
```
